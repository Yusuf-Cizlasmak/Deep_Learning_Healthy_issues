## MODEL EĞİTİMİ 

```python
#Features(Özniteliklerimiz)

x=data.iloc[:,1:].values

y=data.iloc[:,0].values
```

```python
x.shape,y.shape
```

```python
#Train-Test_Split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

esit_mi=len(x_train)/len(y_train) and len(x_test)/len(y_test) # verilerin eşit dağılması


if esit_mi==1:
    print("Verileriniz eşit dağıtılmıştır.")
else :
    print("Verileriniz eşit dağıtılMAmıştır.")
```

```python
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
#vektörizasyon işlemleri 

y_train=y_train.reshape(-1,1)

y_test=y_test.reshape(-1,1)


y_train.shape, y_test.shape
```

### LOGISTIC REGRESSION 

```python
from sklearn.linear_model import LogisticRegression

log=LogisticRegression() #kapsül 

#model eğitimi
model=log.fit(x_train,y_train) 
```

```python
#Tahmin işlemleri 
y_tahmin=log.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_tahmin,y_test))
```

```python

def Accuracy_and_k_cross_val(model,x_train,y_train,cv):

    """
    cross_val_score alacağı 4 önemli parametresi var.
    1. parametre "estimator" = yani bizim durumumuzda model değişkeni

    2.parametre ise X (Neyin fit edildiği)

    3.parametre ise Y (Neye göre fit edildiği)

    4. cv: kaç katlamalı (kaça katlayacağımız)
    """

    from sklearn.model_selection import cross_val_score
    
    accuracies = cross_val_score(estimator = model, X = x_train, y = y_train, cv = cv) 
    print("Modelin başarısı: {:.2f} %".format(accuracies.mean()*100)) #accurayların ortalamasını verir.
    print("Modeller arasındaki standart sapma {:.2f} %".format(accuracies.std()*100))

    print("K-Cross-Validation sonuçları: \n")

    return cross_val_score(estimator=model,X=x_train,y=y_train,cv=cv) # kaç katmana ayırdıysak ona göre veri setini böler ve tüm ihtimalleri değerlendirerek bize bir sonuç verir.

Accuracy_and_k_cross_val(model,x_train,y_train,cv=5)
```

### Confusion Matrix 

```python
from sklearn.metrics import confusion_matrix


cf_matrix=confusion_matrix(y_test,y_tahmin)

fig, ax= plt.subplots(figsize=(15,15))

sns.heatmap(cf_matrix,linewidths=1,annot=True,ax=ax,fmt='g')

plt.show()
```

## K-NN ( En Yakın Komşu) Algoritması 

```python
#? Oldukça çok komşu seçmek bizi KNN algoritmasında daha başarılı sonuçlara götürür mü ? 

from sklearn.neighbors import KNeighborsClassifier

basarilar= []

for k in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train) 
    accuracy=knn.score(x_test,y_test)
    basarilar.append(accuracy)

#Grafiğini 

plt.plot(range(1,21),basarilar)
plt.title("KNN Farklı K değerlerindeki Doğruluk Oranı")
plt.xlabel("Komşu değerleri")
plt.ylabel("Doğruluk Oranı")
plt.show()
```

```python
knn_2=KNeighborsClassifier(n_neighbors=2)

model_knn=knn_2.fit(x_train,y_train) 

y_pred_knn=knn_2.predict(x_test)
```

```python
print(classification_report(y_pred_knn,y_test))
```

```python
Accuracy_and_k_cross_val(model_knn,x_train,y_train,cv=5)
```

```python
cf_matrix=confusion_matrix(y_test,y_pred_knn)

fig, ax= plt.subplots(figsize=(15,15))

sns.heatmap(cf_matrix,linewidths=1,annot=True,ax=ax,fmt='g')

plt.show()
```
